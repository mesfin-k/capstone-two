# 04 - Modeling

This notebook builds and evaluates three models to predict fraudulent transactions, ensuring no data leakage by preprocessing only on training data. The goal is to maximize **recall** (minimize missed fraud) while balancing **F1 score** for imbalanced data (~10% fraud). Models include Logistic Regression (linear, interpretable), Random Forest (tree-based, nonlinear, with Gini and entropy criteria), and XGBoost (gradient boosting, strong for imbalance). Hyperparameter tuning uses cross-validation. We investigate perfect test set scores (recall: 1.0, F1: 1.0) for leakage via feature-target correlations, train-test distribution similarity, time-based split, and holdout set validation.

```python
# Import libraries for modeling, evaluation, and preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay,
    RocCurveDisplay, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
)
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
import joblib
from xgboost import XGBClassifier
```

```python
# Investigate perfect test set scores
# Load raw data
df = pd.read_csv('sentinel_features_for_modeling.csv')

# Encode 'type' for correlation calculation
le = LabelEncoder()
df_encoded = df.copy()
df_encoded['type'] = le.fit_transform(df['type'])

# Select numerical features for correlation (exclude non-predictive identifiers)
numerical_features = ['step', 'log_amount', 'log_oldbalanceOrg', 'log_newbalanceOrig', 
                      'log_oldbalanceDest', 'log_newbalanceDest', 'balance_shift_diff', 
                      'balanced', 'type', 'isFraud']
print("Feature-Target Correlations:\n", df_encoded[numerical_features].corr()['isFraud'].sort_values(ascending=False))

# Preprocess for distribution comparison
features = ['step', 'log_amount', 'log_oldbalanceOrg', 'log_newbalanceOrig', 
            'log_oldbalanceDest', 'log_newbalanceDest', 'balance_shift_diff', 
            'balanced', 'type']
X = df[features]
y = df['isFraud']
X['type'] = le.transform(X['type'])

# Time-based split for distribution comparison
df = df.sort_values('step')
split_idx = int(0.8 * len(df))
X_train = df[features].iloc[:split_idx]
X_test = df[features].iloc[split_idx:]
y_train = df['isFraud'].iloc[:split_idx]
y_test = df['isFraud'].iloc[split_idx:]
X_train['type'] = le.transform(X_train['type'])
X_test['type'] = le.transform(X_test['type'])

# Compare train-test distributions
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_df = pd.DataFrame(X_train_scaled, columns=features)
X_test_df = pd.DataFrame(X_test_scaled, columns=features)
print("\nTrain Mean:\n", X_train_df.mean())
print("\nTest Mean:\n", X_test_df.mean())
print("\nTrain Std:\n", X_train_df.std())
print("\nTest Std:\n", X_test_df.std())
```

```python
# Preprocess with time-based split to avoid temporal leakage
X = df[features]
y = df['isFraud']
X['type'] = le.transform(X['type'])

# Time-based split
split_idx = int(0.8 * len(df))
X_train = X.iloc[:split_idx]
X_test = X.iloc[split_idx:]
y_train = y.iloc[:split_idx]
y_test = y.iloc[split_idx:]

# Scale features (fit on X_train only)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Verify shapes and class balance
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print("Class balance in y_train:\n", pd.Series(y_train).value_counts(normalize=True))
```

```python
# Train Logistic Regression
logreg = LogisticRegression(random_state=42, max_iter=1000, solver='liblinear', class_weight='balanced')
logreg.fit(X_train, y_train)
y_pred_log = logreg.predict(X_test)
y_prob_log = logreg.predict_proba(X_test)[:, 1]

print("\nLogistic Regression Confusion Matrix:\n", confusion_matrix(y_test, y_pred_log))
print("\nLogistic Regression Classification Report:\n", classification_report(y_test, y_pred_log))
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_log = cross_val_score(logreg, X_train, y_train, cv=cv, scoring='f1')
print("\nLogReg Cross-Validation F1 scores:", cv_scores_log)
print("Mean CV F1 (LogReg):", np.mean(cv_scores_log), "\u00b1", np.std(cv_scores_log))

# Train and tune Random Forest
param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'class_weight': ['balanced', None],
    'criterion': ['gini', 'entropy']
}
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=cv, scoring='f1')
grid_rf.fit(X_train, y_train)
tuned_rf = grid_rf.best_estimator_
print("Best RF params:", grid_rf.best_params_)
y_pred_rf = tuned_rf.predict(X_test)
y_prob_rf = tuned_rf.predict_proba(X_test)[:, 1]

print("\nTuned Random Forest Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nTuned Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
cv_scores_rf = cross_val_score(tuned_rf, X_train, y_train, cv=cv, scoring='f1')
print("\nRF Cross-Validation F1 scores:", cv_scores_rf)
print("Mean CV F1 (RF):", np.mean(cv_scores_rf), "\u00b1", np.std(cv_scores_rf))

# Train and tune XGBoost
param_grid_xgb = {
    'n_estimators': [50, 100],
    'max_depth': [3, 6],
    'learning_rate': [0.01, 0.1],
    'scale_pos_weight': [1, len(y_train[y_train==0]) / len(y_train[y_train==1])]
}
grid_xgb = GridSearchCV(XGBClassifier(random_state=42, eval_metric='logloss'), param_grid_xgb, cv=cv, scoring='f1')
grid_xgb.fit(X_train, y_train)
tuned_xgb = grid_xgb.best_estimator_
print("Best XGBoost params:", grid_xgb.best_params_)
y_pred_xgb = tuned_xgb.predict(X_test)
y_prob_xgb = tuned_xgb.predict_proba(X_test)[:, 1]

print("\nTuned XGBoost Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
print("\nTuned XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
cv_scores_xgb = cross_val_score(tuned_xgb, X_train, y_train, cv=cv, scoring='f1')
print("\nXGBoost Cross-Validation F1 scores:", cv_scores_xgb)
print("Mean CV F1 (XGBoost):", np.mean(cv_scores_xgb), "\u00b1", np.std(cv_scores_xgb))
```

```python
# Visualize confusion matrices
models_preds = {
    "Logistic Regression": y_pred_log,
    "Random Forest": y_pred_rf,
    "XGBoost": y_pred_xgb
}
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, (name, preds) in enumerate(models_preds.items()):
    ConfusionMatrixDisplay.from_predictions(y_test, preds, ax=axes[i])
    axes[i].set_title(f"{name} - Confusion Matrix")
plt.tight_layout()
plt.show()

# Plot ROC curves
fig, ax = plt.subplots(figsize=(8, 6))
RocCurveDisplay.from_estimator(logreg, X_test, y_test, ax=ax, name="LogReg")
RocCurveDisplay.from_estimator(tuned_rf, X_test, y_test, ax=ax, name="RF")
RocCurveDisplay.from_estimator(tuned_xgb, X_test, y_test, ax=ax, name="XGBoost")
plt.title("ROC Curves Comparison")
plt.show()
```

```python
# Feature importances for tree-based models
feature_names = ['step', 'log_amount', 'log_oldbalanceOrg', 'log_newbalanceOrig', 
                 'log_oldbalanceDest', 'log_newbalanceDest', 'balance_shift_diff', 
                 'balanced', 'type']
def plot_feature_importances(model, title):
    importances = model.feature_importances_
    indices = np.argsort(importances)[::-1]
    plt.figure(figsize=(10, 6))
    plt.bar(range(len(importances)), importances[indices])
    plt.xticks(range(len(importances)), [feature_names[i] for i in indices], rotation=90)
    plt.title(title)
    plt.show()

plot_feature_importances(tuned_rf, "Random Forest Feature Importances")
plot_feature_importances(tuned_xgb, "XGBoost Feature Importances")
```

```python
# Validate on holdout set (last 10% of step values)
holdout_idx = int(0.9 * len(df))
X_holdout = df[features].iloc[holdout_idx:]
y_holdout = df['isFraud'].iloc[holdout_idx:]
X_holdout['type'] = le.transform(X_holdout['type'])
X_holdout_scaled = scaler.transform(X_holdout)

# Evaluate XGBoost on holdout set
y_pred_holdout = tuned_xgb.predict(X_holdout_scaled)
y_prob_holdout = tuned_xgb.predict_proba(X_holdout_scaled)[:, 1]
print("\nXGBoost Holdout Set Confusion Matrix:\n", confusion_matrix(y_holdout, y_pred_holdout))
print("\nXGBoost Holdout Set Classification Report:\n", classification_report(y_holdout, y_pred_holdout))
print("XGBoost Holdout ROC-AUC:", roc_auc_score(y_holdout, y_prob_holdout))
```

```python
# Re-run without suspect features if leakage is confirmed
# If correlations show balanced or balance_shift_diff ~Â±1, exclude them
features_no_leakage = ['step', 'log_amount', 'log_oldbalanceOrg', 'log_newbalanceOrig', 
                       'log_oldbalanceDest', 'log_newbalanceDest', 'type']
X_no_leakage = df[features_no_leakage]
X_no_leakage['type'] = le.transform(X_no_leakage['type'])

# Time-based split
X_train_nl = X_no_leakage.iloc[:split_idx]
X_test_nl = X_no_leakage.iloc[split_idx:]
scaler_nl = StandardScaler()
X_train_nl = scaler_nl.fit_transform(X_train_nl)
X_test_nl = scaler_nl.transform(X_test_nl)

# Re-train XGBoost
grid_xgb.fit(X_train_nl, y_train)
tuned_xgb_nl = grid_xgb.best_estimator_
y_pred_xgb_nl = tuned_xgb_nl.predict(X_test_nl)
y_prob_xgb_nl = tuned_xgb_nl.predict_proba(X_test_nl)[:, 1]
print("\nXGBoost (No Leakage Features) Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb_nl))
print("\nXGBoost (No Leakage Features) Classification Report:\n", classification_report(y_test, y_pred_xgb_nl))
```

```python
# Apply SMOTE if recall < 0.85
if recall_score(y_test, y_pred_xgb_nl) < 0.85:
    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
    grid_xgb.fit(X_train_smote, y_train_smote)
    tuned_xgb_smote = grid_xgb.best_estimator_
    y_pred_xgb_smote = tuned_xgb_smote.predict(X_test)
    y_prob_xgb_smote = tuned_xgb_smote.predict_proba(X_test)[:, 1]
    print("\nXGBoost (SMOTE) Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb_smote))
    print("\nXGBoost (SMOTE) Classification Report:\n", classification_report(y_test, y_pred_xgb_smote))

# Optional: Threshold tuning if recall still < 0.85
    if recall_score(y_test, y_pred_xgb_smote) < 0.85:
        thresholds = np.arange(0.1, 1.0, 0.1)
        best_recall = 0
        best_threshold = 0.5
        for thresh in thresholds:
            y_pred_thresh = (y_prob_xgb_smote >= thresh).astype(int)
            recall = recall_score(y_test, y_pred_thresh)
            if recall > best_recall:
                best_recall = recall
                best_threshold = thresh
        print(f"\nBest threshold: {best_threshold}, Recall: {best_recall}")
        y_pred_xgb_thresh = (y_prob_xgb_smote >= best_threshold).astype(int)
        print("XGBoost (SMOTE + Threshold) Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb_thresh))
        print("XGBoost (SMOTE + Threshold) Classification Report:\n", classification_report(y_test, y_pred_xgb_thresh))
```

```python
# Compare models (original features, time-based split)
metrics = {}
for name, y_pred, y_prob in [("LogReg", y_pred_log, y_prob_log), ("RF", y_pred_rf, y_prob_rf), ("XGBoost", y_pred_xgb, y_prob_xgb)]:
    metrics[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_prob)
    }
df_metrics = pd.DataFrame(metrics).T.sort_values(by='Recall', ascending=False)
print("\nModel Comparison Table (sorted by Recall):")
print(df_metrics)
```

# Final Model Recommendation (Revised)

## Focus

- **Recall** is prioritized to minimize **missed fraud (false negatives)**, which is more costly than false positives in fraud detection.
- **F1 Score** balances recall and precision, important in imbalanced datasets (~10% fraud).
- **ROC-AUC** assesses overall class separation power.

---

## Recommended Model

- **Primary Model:** â **XGBoost**, trained without leaky features and tuned with `scale_pos_weight = 9.0`.
- **Alternative (if interpretability/scalability is preferred):** Random Forest with `class_weight='balanced'` and `criterion='entropy'`.
- **If leakage is confirmed:** Use **no-leakage feature version** of XGBoost.
- **If holdout recall < 0.85:** Consider using **SMOTE-enhanced** or **threshold-tuned** XGBoost.

---

## Justification

---

### Scalability

- **XGBoost** supports early stopping and handles large datasets well.
- **Random Forest** scales well in distributed environments.
- **Logistic Regression** is scalable but lacks fraud-detection strength.

---

### Maintenance Plan

- Retrain quarterly or on drift.
