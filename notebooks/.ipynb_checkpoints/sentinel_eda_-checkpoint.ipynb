```python
# === 5. Univariate Analysis (Optional) ===
# Purpose: Understand overall distributions of key features
# Why: Helps detect skew or need for transformations

# Example: Distribution of transaction amounts
plt.figure(figsize=(8, 5))
sns.histplot(df['amount'], bins=50, kde=True)
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Observation:
# Does the amount distribution show unusual spikes or a long tail? Are there very large transactions?
# Large or rare amounts may be more likely linked to fraud.

# === Violin Plot for Transaction Amount ===
plt.figure(figsize=(8, 6))
sns.violinplot(x='isFraud', y='amount', data=df)
plt.yscale("log")
plt.title("Violin Plot: Transaction Amount by Fraud Status")
plt.xlabel("Is Fraud")
plt.ylabel("Amount")
plt.tight_layout()
plt.show()

# Observation:
# Violin plot shows full distribution shape  helps visualize skew and outliers in a richer way than boxplot.

# ===Fraud Count by Step (Time) ===
fraud_by_step = df[df['isFraud'] == 1].groupby('step').size()

plt.figure(figsize=(10, 6))
fraud_by_step.plot(kind='line', marker='o')
plt.title("Fraudulent Transactions Over Time (Step)")
plt.xlabel("Step")
plt.ylabel("Number of Fraud Cases")
plt.grid(True)
plt.tight_layout()
plt.show()

# Observation:
# Useful to detect if fraud happens more at certain times  this informs modeling and feature engineering.

# === Scatterplot Amount vs Step Colored by Fraud ===
plt.figure(figsize=(10, 6))
sns.scatterplot(x='step', y='amount', hue='isFraud', data=df, alpha=0.7)
plt.yscale("log")
plt.title("Amount vs Step, Colored by Fraud")
plt.xlabel("Step")
plt.ylabel("Amount")
plt.legend(title='Is Fraud')
plt.tight_layout()
plt.show()

#=== Zero Balance Check

print("Sender zero balance %:", (df['oldbalanceOrg'] == 0).mean() * 100)
print("Recipient zero balance %:", (df['oldbalanceDest'] == 0).mean() * 100)

sns.barplot(x='isFraud', y=(df['oldbalanceOrg'] == 0).astype(int), data=df)
plt.title("Sender Zero Balance vs Fraud")
plt.show()

sns.barplot(x='isFraud', y=(df['oldbalanceDest'] == 0).astype(int), data=df)
plt.title("Recipient Zero Balance vs Fraud")
plt.show()


```
